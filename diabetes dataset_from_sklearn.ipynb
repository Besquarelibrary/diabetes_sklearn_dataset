```python
#######ONE HOT ENCODING###########
#one hot encoding will apply for the unordered values in a columns 
#label encoding will apply for the ordered values in a column
#encoding will be used when the all the the coulumns have the numerical data and some having the naming data to compare and relate those columns inorder to predict the efficient data we use this encoding method
import pandas as pd
champion_stats={'Team':['India','Australia','West Indis','Pakistan','Srilanka'],
               'ICC_rank':[2,3,7,8,4],
               'World_Champion_year':[2011,2015,1979,1992,1996],
               'Points':[874,787,753,673,855]}
match_stats={'Team':['India','Australia','West Indis','Pakistan','Srilanka'],
            'world_cup_played':[11,10,11,9,8],
            'ODIs_played':[733,988,712,679,662]}
df1=pd.DataFrame(champion_stats)
df2=pd.DataFrame(match_stats)
df1
```

```python
#linear regression model will drop the dummy variables automatically 
#we can also remove the dummies from the dataframe by using drop 
drop_result=dataframe_aftermerge.drop(['Team','Pakistan'],axis='columns')
#here using drop method we are removing list of columns from the merged dataframe and axis should be column
```
